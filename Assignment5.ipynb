# Question:1

```python
age=int(input("enter a person age:"))
if age>=0 and age<=12:
    print("Child")
elif age>=13 and age<=19:
    print("Teenager")
elif age>=20 and age<=64:
    print("Adult")
elif age>=65:
    print("Senior")
else:
    print("Invalid age")
```
